```python
# Specify the name of the city
city = "Augsburg, Germany"

# Get the boundary of Augsburg as a GeoDataFrame
city_gdf = ox.geocode_to_gdf(city)

# Use the boundary to get green spaces within Augsburg
tags = {"leisure": "park"}
green_spaces = ox.features_from_place(city, tags=tags)
green_spaces = green_spaces.dropna(axis='columns', how='all')
green_spaces = green_spaces[green_spaces.geometry.type.isin(['Polygon'])]
green_spaces = green_spaces[['leisure', 'geometry']]
green_spaces = green_spaces.rename(columns={'leisure': 'class'})
green_spaces = green_spaces.reset_index()
green_spaces = green_spaces.drop('element_type', axis=1)
green_spaces
```

```python
tags = {"leisure": "playground"}
playground_spaces = ox.features_from_place(city, tags=tags)
playground_spaces = playground_spaces.dropna(axis='columns', how='all')
playground_spaces = playground_spaces[playground_spaces.geometry.type.isin(['Polygon'])]
playground_spaces = playground_spaces[['leisure', 'geometry']]
playground_spaces = playground_spaces.rename(columns={'leisure': 'class'})
playground_spaces = playground_spaces.reset_index()
playground_spaces = playground_spaces.drop('element_type', axis=1)
playground_spaces
```

```python
tags = {"landuse": "grass"}
grass_spaces = ox.features_from_place(city, tags=tags)
grass_spaces = grass_spaces.dropna(axis='columns', how='all')
grass_spaces = grass_spaces[grass_spaces.geometry.type.isin(['Polygon'])]
grass_spaces = grass_spaces[['landuse', 'geometry']]
grass_spaces = grass_spaces.rename(columns={'landuse': 'class'})
grass_spaces = grass_spaces.reset_index()
grass_spaces = grass_spaces.drop('element_type', axis=1)
grass_spaces
```

```python
tags = {"landuse": "recreation_ground"}
recreation_ground_spaces = ox.features_from_place(city, tags=tags)
recreation_ground_spaces = recreation_ground_spaces.dropna(axis='columns', how='all')
recreation_ground_spaces = recreation_ground_spaces[recreation_ground_spaces.geometry.type.isin(['Polygon'])]
recreation_ground_spaces = recreation_ground_spaces[['landuse', 'geometry']]
recreation_ground_spaces = recreation_ground_spaces.rename(columns={'landuse': 'class'})
recreation_ground_spaces = recreation_ground_spaces.reset_index()
recreation_ground_spaces = recreation_ground_spaces.drop('element_type', axis=1)
recreation_ground_spaces
```

```python
tags = {"building": True}
building_spaces = ox.features_from_place(city, tags=tags)
building_spaces = building_spaces.sample(n=3000, random_state=42)
building_spaces = building_spaces.dropna(axis='columns', how='all')
building_spaces = building_spaces[building_spaces.geometry.type.isin(['Polygon'])]
building_spaces = building_spaces[['building', 'geometry']]
building_spaces = building_spaces.rename(columns={'building': 'class'})
building_spaces['class'] = 'building'
building_spaces = building_spaces.reset_index()
building_spaces = building_spaces.drop('element_type', axis=1)
building_spaces
```

```python
import pandas as pd

gdf_list = [recreation_ground_spaces, grass_spaces, cemetery_spaces, memorial_spaces, funeral_hall_spaces, playground_spaces, green_spaces, building_spaces]

# Concatenating all GeoDataFrames
combined_gdf = pd.concat(gdf_list, ignore_index=True)
combined_gdf
```

```python
combined_gdf.to_file("data/osm_polygon.shp")
combined_gdf.to_file("data/osm_polygon.geojson", driver='GeoJSON')
combined_gdf.to_file("data/osm_polygon.gpkg", driver="GPKG")
combined_gdf.to_parquet("data/osm_polygon.parquet")
```
